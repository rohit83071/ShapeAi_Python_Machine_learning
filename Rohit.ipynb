we never train the model on all the  data that we have ,we always make sure to atleast have a test dataset ,which is different from training dataset

```python
from sklearn.model_selection import train_test_split
X = boston.drop('Price',axis=1)
Y = boston['Price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=1)

print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

 now let import linear recursion model from sklearn and train it on training dataset
 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
print(lin_model)
```

```python
Y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, Y_train_predict)))
print("The Model Performance For Training set ")
print('RMSE is{}'.format(rmse))
print("\n")

#on training test
Y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, Y_test_predict)))


print("the model performance for training test")
print('RMSE is{}'.format(rmse))
print("\n")
```
